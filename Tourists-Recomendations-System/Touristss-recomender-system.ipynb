```python
ratings_with_name=ratings.merge(tourismid,on='Place_Id')
```

```python
ratings_with_name
```

```python
num_rating_df=ratings_with_name.groupby('Place_Name').count()['Place_Ratings'].reset_index()
num_rating_df.rename(columns={'Place_Ratings':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df=ratings_with_name.groupby('Place_Name').mean()['Place_Ratings'].reset_index()
avg_rating_df.rename(columns={'Place_Ratings':'avg_ratings'},inplace=True)
avg_rating_df
```

```python
popular_df=num_rating_df.merge(avg_rating_df,on='Place_Name')
```

```python
popular_df
```

```python
popular_df[popular_df['num_ratings']>=15].sort_values('avg_ratings',ascending=False).round(2).head(20)
```

```python
popular_df.merge(tourismid,on='Place_Name').drop_duplicates('Place_Name')[['Place_Name','City','Price','num_ratings','avg_ratings']].head(20)
```

```python
popular_df
```

```python
x=ratings_with_name.groupby('User_Id').count()['Place_Ratings']>40
padhe_lekhe_users=x[x].index



```

```python
filtered_rating=ratings_with_name[ratings_with_name['User_Id'].isin(padhe_lekhe_users)]
```

```python
y=filtered_rating.groupby('Place_Name').count()['Place_Ratings']>=7
famous_places=y[y].index





```

```python
final_ratings=filtered_rating[filtered_rating['Place_Name'].isin(famous_places)]
```

```python
pt=final_ratings.pivot_table(index='Place_Name',columns='User_Id',values='Place_Ratings')
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
    from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores=cosine_similarity(pt)
```

```python
similarity_scores
```

```python
def recommend(place_name):
    index=np.where(pt.index==place_name)[0][0]
    similar_items=sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    
    for i in similar_items:
       print(pt.index[i[0]])
        
    
```

```python
recommend('Taman Srigunting')
```
